### Preprocessing

```python
# rearranging the columns in df
df = df[["Date", "Time", "Name", "Chat"]]
df.head()
```

```python
# remove '[' in Date col
Message1 = df['Date'].str.split("[", n=1,expand=True)
Message1.head()

df["Date"] = Message1[1]
df.head()
```

```python
def sentimentalAnalysis(data, columnname):
    # downloading vader_lexicon for process
    import matplotlib.pyplot as plt
    import seaborn as sns
    # Importing necessary package
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    import nltk
    nltk.downloader.download('vader_lexicon')
    sid = SentimentIntensityAnalyzer()
    
    # Removing null entries
    data.dropna(inplace=True)
    
    # Checing for comments
    # senti = sid.polarity_scores(dataset['Chat'][4])
    
    # Creating respective column
    data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText)) # .apply() -> apply the process for all rows
    data['compound'] = data['scores'].apply(lambda score_dict: score_dict['compound'])
    data['Negative'] = data['scores'].apply(lambda score_dict: score_dict['neg'])
    data['Positive'] = data['scores'].apply(lambda score_dict: score_dict['pos'])
    data['Neutral'] = data['scores'].apply(lambda score_dict: score_dict['neu'])
    
    # Creating final pos, neg using compound score
    data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
    plt.clf()
    # comp = sns.countplot(x='comp_score', hue='Name', data=data, palette='magma')
    # comp.figure.savefif("data_charts.png")
    
    # checking how many pos, neg
    posneg = pd.DataFrame(data['comp_score'].value_counts())
    return posneg, data
```

```python
pos, dataSenti = sentimentalAnalysis(dataset, columnname='Chat')
dataSenti
```

### Topic modeling

#### 1. Using TF-TDF vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# preprocessing
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm=tfidf.fit_transform(df["Chat"]) # will create model and with same input, it will predict
```

#### Using NMF

```python
# topic modeling using NMF
from sklearn.decomposition import NMF
nmf_model = NMF(n_components=5, random_state=42)
nmf_model.fit(dtm)
```

```python
for index, topic in enumerate(nmf_model.components_):
    results = ([tfidf.get_feature_names()[i] for i in topic.argsort()[-10:]]) # get 10 from each topic, if need to remove few stop_words - refer internet
    print(results) # result is unsupervised
```

```python
topic_results = nmf_model.transform(dtm)
df["Topic_using_NMF"] = topic_results.argmax(axis=1)
```

```python
df["Topic_using_NMF"].value_counts()
```

#### 2. Using Countervectorizer
